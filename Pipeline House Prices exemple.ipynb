```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df
```

```python
# target
y = df['SalePrice']
# features
X = df[['LotArea','Street','MiscFeature']]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn import set_config
set_config(display='diagram')
```

```python
pipe_category = Pipeline([
    ('replace_NaN2',SimpleImputer(strategy='most_frequent')),
    ('pipeline_cat',OneHotEncoder())])
pipe_category
```

```python
pipe_numeric = Pipeline([('replace_NaN',SimpleImputer()),('minmaxscaler',MinMaxScaler())])
pipe_numeric
```

```python
from sklearn.compose import ColumnTransformer
```

```python
pipe_transf = ColumnTransformer([('transf_cat', pipe_category, df.dtypes=='object'),
                                 ('transf_num', pipe_numeric, df.dtypes=='int64'),
                                 ('transf_float', pipe_numeric, df.dtypes=='float64')])
pipe_transf
```

```python
# pipe_transf.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
pipe_model = Pipeline([('pipe_transf',pipe_transf),
                       ('modele',LinearRegression())])
pipe_model
```

```python
# pipe_model.fit(X,y)
```

```python
pipe_model.score(X,y)
```
